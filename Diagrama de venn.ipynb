```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
import pandas as pd
print(matplotlib.__version__)
```

```python
# Define the path to the text file
path_52 = "C://Users//andre//Documents//Mestrado Bioinformática//Mestrado//2º Semestre//OMICs//Aulas praticas//Projecto//anno.gr_52.txt"
path_53 = "C://Users//andre//Documents//Mestrado Bioinformática//Mestrado//2º Semestre//OMICs//Aulas praticas//Projecto//anno.gr_53.txt"
path_both = "C://Users//andre//Documents//Mestrado Bioinformática//Mestrado//2º Semestre//OMICs//Aulas praticas//Projecto//anno.both.txt"
path_DEGs = "C://Users//andre//Documents//Mestrado Bioinformática//Mestrado//2º Semestre//OMICs//Aulas praticas//Projecto//DEGs.txt"

# Read the text files and convert their to lists
with open(path_52, 'r') as file:
    anno_52 = [line.strip() for line in file if line.strip()]

with open(path_53, 'r') as file:
    anno_53 = [line.strip() for line in file if line.strip()]

with open(path_both, 'r') as file:
    anno_both = [line.strip() for line in file if line.strip()]

with open(path_DEGs, 'r') as file:
    DEGs = [line.strip() for line in file if line.strip()]
```

Sample genes with ChIP-seq peaks

```python
# Define the sets genes in each condition
set_anno_52 = set(anno_52)
set_anno_53 = set(anno_53)

# Venn Diagram
venn_diagram = venn2(subsets=(set_anno_52, set_anno_53), set_labels=('ChIP Replicate 1', 'ChIP Replicate 2'))

venn_diagram.get_patch_by_id('10').set_color('#A6CEE3') 
venn_diagram.get_patch_by_id('01').set_color('#FB9A99')  
venn_diagram.get_patch_by_id('11').set_color('#FDBF6F')

# Customize borders
for subset in ('10', '01', '11'):
    venn_diagram.get_patch_by_id(subset).set_edgecolor('black')
    venn_diagram.get_patch_by_id(subset).set_linewidth(1)

# Customize labels
for text in venn_diagram.subset_labels:
    if text: 
        text.set_fontsize(12)
        #text.set_fontweight('bold')

# Add title
plt.title("Sample genes with ChIP-seq peaks", fontsize=14)

# Show graphic
plt.show()
```

ChIP-seq Peak Localization

```python
number2 = [413, 344, 55]
types = ('Upstream', 'Overlap Start', 'Inside')

barplot = pd.DataFrame({'Number Genes': number2, 'Localization': types})

color = [(102/255, 194/255, 165/255, 0.7), 
         (141/255, 160/255, 203/255, 0.7),  
         (166/255, 216/255, 84/255, 0.7)] 

# Create bars horizontally
fig, ax = plt.subplots()
barsh = ax.barh(barplot['Localization'], barplot['Number Genes'],color=color)

# Custom X axis
plt.yticks(fontsize=12, fontweight='bold')
plt.gca().xaxis.set_tick_params(pad=7)

# Remove top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Labeling
plt.xlabel("Number of Genes", fontweight='bold') 

# Adjust the positions of y axis title
plt.gca().yaxis.set_label_coords(-0.1, 0.5)

# Set custom ticks and labels for the y-axis
plt.xticks([0, 100, 200, 300, 400, 500])

# Add labels to the bars
for i, bar in enumerate(barsh):
    plt.text(bar.get_width() /2, bar.get_y() + bar.get_height() / 2, str(barplot['Number Genes'][i]), ha='center', va='center', fontweight='bold', color='black')

# Show graph
plt.show()
```
